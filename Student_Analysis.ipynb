```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Explanation:

pandas → Data manipulation and analysis.
matplotlib and seaborn → Data visualization.

```python
# Load the dataset
data = pd.read_csv('student-mat.csv', sep=';')  # Use the appropriate file name

# Display the first few rows of the dataset
data.head()
```

Explanation:

Loads the dataset into a DataFrame.
head() displays the first 5 rows.

```python
data.info()
```

```python
# Display data types of  column
print("Shape of Dataset:", data.shape)
```

```python
# Display data types
print(data.dtypes)
```

Explanation:

shape shows rows and columns.
dtypes shows data types.
info() gives complete dataset information.

```python
# Summary statistics
data.describe()
```

```python
# check for missing values
data.isnull().sum()
```

Explanation:

isnull().sum()  -  Give the missing values

```python
# Check for duplicates
data.duplicated().sum()

# Remove duplicates if any
data = data.drop_duplicates()
```

Explanation:

duplicated().sum()   Check the duplicate values
drop_duplicates()   Remove the duplicate values

```python
# Average Final Grade (G3)
avg_grade = data['G3'].mean()
print("Average Final Grade:", round(avg_grade, 2))
```

```python
# Students Scoring Above 15
above_15 = data[data['G3'] > 15]

print("Students Scoring Above 15:", len(above_15))
```

```python
# Correlation between Study Time and Final Grade
correlation = data['studytime'].corr(data['G3'])

print("Correlation between Study Time and G3:", round(correlation, 3))
```

```python
# Gender Performance
gender_performance = data.groupby('sex')['G3'].mean()

print(gender_performance)
```

```python
# Distribution of Final Grades
plt.figure(figsize=(8,5))
plt.hist(data['G3'], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Study Time vs Final Grade
plt.figure(figsize=(8,5))
plt.scatter(data['studytime'], data['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
# Average Grade by Gender
gender_avg = data.groupby('sex')['G3'].mean()

plt.figure(figsize=(6,4))
gender_avg.plot(kind='bar')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.show()
```
